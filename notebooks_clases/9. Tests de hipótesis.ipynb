# Inferencia estadística y pruebas de hipótesis

Este notebook tiene como objetivo explicar, de manera práctica, conceptos de **inferencia estadística** y algunas de sus aplicaciones: pruebas de hipótesis para la media, varianza, diferencia de medias.

## 1. Prueba de hipótesis para la media (n grande)
Cuando tenemos un tamaño de muestra grande (n > 30), podemos usar la **distribución normal**.
Ejemplo: ¿el peso medio de los estudiantes es 70 kg?

```python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

# Simulamos datos
datos = np.random.normal(72, 5, 50)

mu0 = 70
media_muestra = np.mean(datos)
std_muestra = np.std(datos, ddof=1)
n = len(datos)

z = (media_muestra - mu0) / (std_muestra/np.sqrt(n))
p_valor = 1 - stats.norm.cdf(z)

print("--- Prueba de hipótesis para la media ---")
print(f"Media muestral: {media_muestra:.2f}")
print(f"Estadístico z: {z:.2f}, p-valor: {p_valor:.4f}")
```

## 2. Prueba t para la media con pocos datos
Cuando n < 30 y no conocemos sigma, usamos la **t de Student**.

```python

datos_peq = np.random.normal(72, 5, 12)
media_peq = np.mean(datos_peq)
t_stat, p_valor_t = stats.ttest_1samp(datos_peq, 70)

print("\n--- Prueba t para la media (n pequeño) ---")
print(f"Media muestral: {media_peq:.2f}")
print(f"t: {t_stat:.2f}, p-valor: {p_valor_t:.4f}")
```

## 3. Diferencia de medias (muestras independientes)
Para comparar dos grupos independientes usamos **t de Student**.

```python

grupo1 = np.random.normal(72, 5, 30)
grupo2 = np.random.normal(75, 5, 30)

stat, p = stats.ttest_ind(grupo1, grupo2)
print("\n--- Diferencia de medias (independientes) ---")
print(f"Media grupo1: {np.mean(grupo1):.2f}, Media grupo2: {np.mean(grupo2):.2f}")
print(f"t: {stat:.2f}, p-valor: {p:.4f}")
```

## 4. Prueba sobre varianza (chi-cuadrado)
Para hipótesis sobre varianza usamos la **chi-cuadrado**.

```python

var_muestra = np.var(grupo1, ddof=1)
sigma0 = 25
chi2 = (len(grupo1)-1) * var_muestra / sigma0
p_var = 1 - stats.chi2.cdf(chi2, df=len(grupo1)-1)

print("\n--- Prueba sobre varianza ---")
print(f"Varianza muestral: {var_muestra:.2f}, chi2: {chi2:.2f}, p-valor: {p_var:.4f}")
```
