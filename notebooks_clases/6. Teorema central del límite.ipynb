## Teorema central del límite

**Idea**: La suma de muchas variables aleatorias independientes tiende a comportarse como una distribución normal, sin importar la distribución original.

**Ejemplo simple**: Generamos 10000 muestras de 30 datos c/u con distribución exponencial(λ=1). Promediamos los valores de cada muestra y vemos cómo esa distribución tiende a ser normal, con media igual a la media original y varianza igual a 𝜎2/𝑛.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generamos 10000 muestras de 30 datos c/u
# con distribución exponencial(λ=1)

n_muestras = 10000
tamano_muestra = 30
muestras = np.random.exponential(1, (n_muestras, tamano_muestra))

promedios_muestrales = np.mean(muestras, axis=1)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(muestras[0], bins=20, alpha=0.7, color='gray')
plt.title('Una muestra con la distribución original\n(Exponencial)')

plt.subplot(1,2,2)
plt.hist(promedios_muestrales, bins=30, alpha=0.7, color='lightblue', density=True)
plt.title(f'Promedios de {tamano_muestra} muestras\n(Se parece a normal)')
plt.show()
```
