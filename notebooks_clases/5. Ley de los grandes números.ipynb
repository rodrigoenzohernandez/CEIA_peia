## Ley de los grandes números

**Idea**: Cuando repetimos un experimento aleatorio muchas veces, el promedio de los resultados se acerca a la “media verdadera”.

**Ejemplo simple**: Lanzamos una moneda. Si la tiramos 10 veces, quizá obtengamos 60 % caras. Si la tiramos 1.000 veces, estaremos mucho más cerca del 50 %.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simular promedios de una moneda (p=0.5)
n = 1000  # número de lanzamientos
lanzamientos = np.random.binomial(1, 0.5, n)
promedios = np.cumsum(lanzamientos) / np.arange(1, n+1)

plt.figure(figsize=(8,4))
plt.plot(promedios, linewidth=1)
plt.axhline(y=0.5, color='r', linestyle='--', label='Valor real (0.5)')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Promedio acumulado')
plt.title('Ley de los grandes números')
plt.legend()
plt.grid(True)
plt.show()
```
